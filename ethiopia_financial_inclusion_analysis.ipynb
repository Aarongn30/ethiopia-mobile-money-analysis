```python
import pandas as pd

df = pd.read_csv("ethiopia_financial_data.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

df = df.set_index('Indicator').T  # Transpose and set index
df.plot(marker='o', figsize=(10,6))
plt.title("Ethiopia Financial Inclusion Trends (2011–2021)")
plt.xlabel("Year")
plt.ylabel("% of Population")
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data for prediction
years = df.index.astype(int).values.reshape(-1, 1)  # X values (2011, 2014, ...)
future_years = np.array([2029, 2030]).reshape(-1, 1)

# Predict for each indicator
predictions = {}

for column in df.columns:
    y = df[column].values
    model = LinearRegression().fit(years, y)
    y_pred = model.predict(future_years)
    predictions[column] = y_pred

# Display predictions
for indicator, values in predictions.items():
    print(f"{indicator} (2029): {values[0]:.2f}%, (2030): {values[1]:.2f}%")
```
